###  Lists

**Lists** are one of the most powerful and most used data structures in Python.  Lists are like supercharged strings in that they are an **ordered sequence** of objects.  A list can be indexed and sliced just like strings so you can pull out subgroups of objects.  In fact, list indexing follows the same rules as string indexing.  However, there is one big difference with lists.  Lists are dequences of any types of objects, not just characters.  Lists could contain integers, floating point values, strings, or even other lists.  Also, an important features of lists is that all the elements of a list do not have to be the same type.  They can be any mixture of types because lists are just that flexible.

Lists are declared using [ ] (square brackets) and when you want to access an element you also use [ ] (just like strings).

Try the following code.  Note how the lists are handled by the print statement.

```Python

print(my_list)
my_list=[1,2,3]
```

###  Dictionaries

**Dictionaries** are another type of python data structure.  These are sequences of objects, but they are **unordered**.   There are no numerical indices for a dictionary.  The way elements are accessed in a dictionary is with a key.  This is a value associated with another value.   When you define a dictionary, you create **key/value** pairs.  A value is then accessed with its key.  Dictionaries are defined with { } but again indexed with [ ].  Try the following code:

```Python
d={'one':1 , 'two':2}
print(d)
print(d['two'])
```

###  String Formatting

Sometimes when printing outputs, you may have to print similar messsages, but just need to change certain values.  You could always build these strings each time with concatenation, but there is an easier way to do this, **string formatting**.  This allows you to put a placeholder in a string and then fill it from a variable.  Let’s look at simple example.  The following uses the string formatting command to build and print an output string.   Take note of the place holder in the string.  Placeholders always start with a % and are followed by the type of variable to be used to build the string (e.g. d for integers, f for floating point, and s for string.  The variable to be used in the string is provided after the string format is defined.  Try the following.

```Python
i = 9
print("The number is %d" % i)
```

````python
#
# Enter your code here
#
````

What if you need to put multiple values in the string?  You can do that as well; you just must add additional placeholders in the format string and pass it multiple variables.  The trick with this is that you must pass the multiple values in as a **tuple**.  Try the following code.

```Python
i = 9
f=3.0
s='Sam'
print("%s the numbers are %d and %f" % (s, i, f) )
```

String formatting gives you lots of options for how you want the output format to look.   These are done with parameters between the % and the field code (e.g. s, d, f, etc.).   An integer number between the % and field code indicates the width of the field.  It will cause the placeholder to be filled with characters of that length exactly, no matter what the input value.   If the placeholder is for a floating-point number, the whole number part will be the length of the field and the decimal part indicates how many figures after the decimal will be printed.
